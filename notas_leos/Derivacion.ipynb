# Derivaciòn numèrica

```julia
using Plots
```

```julia
function D_d(f,x0,h) 
    return (f(x0) - f(x0-h))/h
end
```

```julia
function D_s(f,x0,h) 
    return (f(x0+h) - f(x0-h))/2h
end
```

```julia
function D_c(f,x0,h)
    return imag(f(complex(x0,h)))/h
end
```

```julia
f(x) = x^2 -x^3
```

```julia
Df(x) = 2*x-3*x^2
```

```julia
x = -5.0:0.01:5.0
h = 1e-6
```

```julia
plot(x,abs.(Df.(x) .- D_d.(f,x,h)), yscale=:log10, label ="Derivada derecha")
plot!(x,abs.(Df.(x) .- D_s.(f,x,h)), yscale=:log10, label = "Derivada simètrica")
plot!(x,abs.(Df.(x) .- D_c.(f,x,h)), yscale=:log10, label = "Derivada compleja")
title!("Errores en mètodos de derivaciòn de f(x)")
```
